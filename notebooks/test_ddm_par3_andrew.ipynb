```python
from ssms import Simulator 

ddm_par3 = Simulator("ddm_par3")
ddm_par3_out = ddm_par3.simulate(theta={'vh': 2, 'vl1': 1, 
                                'vl2': 0.5, 'coh':1, 'a': 1, 'zh':0.5, 'zl1':0.5, 'zl2': 0.5, 
                                't': 0.3}, n_samples=1000)
#  print(ddm_par3_out.keys())
# # ddm_par3_sim = simulator(model='ddm_par3', theta={'vh': 2, 'vl1': 1, 
#                                               'vl2': 0.5, 'a': 1, 'zh':0.5, 'zl1':0.5, 'zl2': 0.5, 
#                                               't': 0.3}, n_samples=1000)
plt.hist(
    ddm_par3_out["rts"][ddm_par3_out["rts"] != -999] * ddm_par3_out["choices"][ddm_par3_out["rts"] != -999],
    bins=50,
    histtype="step",
    color="black",
    label="Weibull Deadline",
    density=True,
)
plt.xlabel("Reaction Times (neg RTs are errors)")
plt.ylabel("Density")
```
