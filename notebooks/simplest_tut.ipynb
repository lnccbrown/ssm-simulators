```python
# Import the Simulator class (also available as ssms.Simulator)
from ssms import Simulator

# Create a simulator for the DDM model
ddm_sim = Simulator("ddm")

# Run a simulation with named parameters
sim_out_class = ddm_sim.simulate(
    theta={"v": -1., "a": 1.5, "z": 0.75, "t": 0.25}, 
    n_samples=100000,
    n_threads=12,
)

# The output format is identical to the functional approach
print("Keys:", sim_out_class.keys())
print("RT shape:", sim_out_class["rts"].shape)
print("Choices shape:", sim_out_class["choices"].shape)
```

```python
from matplotlib import pyplot as plt 

plt.hist(sim_out_class["rts"] * sim_out_class["choices"], bins=100, density=True)
```
